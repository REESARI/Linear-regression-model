```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (20.0,10.0)

data = pd.read_csv('headbrain.csv')
print(data.shape)
data.head()
```

```python
X = data['Head Size(cm^3)'].values
Y = data['Brain Weight(grams)'].values
```

```python
# Mean X and Y
mean_x = np.mean(X)
mean_y = np.mean(Y)

# Total number of values
n = len(X)

# Using the formula to calculate b1 and b0
numer = 0
denom = 0
for i in range(n):
    numer += (X[i] - mean_x) * (Y[i] - mean_y)
    denom += (X[i] - mean_x) ** 2

b1 = numer / denom
b0 = mean_y - (b1 * mean_x)

# Print coefficients
print(b0, b1)
```

```python
# Scatter Plot of Original Data
plt.scatter(X, Y, color='orange', label='Scatter Plot')

# Predicted Line Equation
Y_pred = b0 + b1 * X

# Plotting Regression Line
plt.plot(X, Y_pred, color='green', label='Regression Line')

plt.xlabel('Head Size in cm³')
plt.ylabel('Brain Weight in grams')
plt.legend()
plt.show()
```

```python
# R2 Score Calculation
ss_tot = 0
ss_res = 0

for i in range(n):
    ss_tot += (Y[i] - mean_y) ** 2  # Total Sum of Squares
    ss_res += (Y[i] - Y_pred[i]) ** 2  # Residual Sum of Squares

r2 = 1 - (ss_res / ss_tot)

print(f"R² Score: {r2}")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Data Points
X = np.array([1, 2, 3, 4, 5])
Y = np.array([5, 7, 9, 11, 13])

# Reshape X to 2D Array (Mandatory for Sklearn)
X = X.reshape(len(X), 1)

# Creating Model
reg = LinearRegression()

# Fitting Data
reg.fit(X, Y)

# Predicting Y
Y_pred = reg.predict(X)

# Calculating MSE, RMSE, and R2 Score
mse = mean_squared_error(Y, Y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(Y, Y_pred)

# Printing Results
print("RMSE:", rmse)
print("R² Score:", r2)
```
